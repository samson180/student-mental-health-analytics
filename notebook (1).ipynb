```python
# Install duckdb to run SQL in Python
!pip install duckdb --quiet

import duckdb
import pandas as pd

# Read the file directly from your uploaded files panel
students_df = pd.read_csv("/content/Student Mental health.csv")

# Connect to DuckDB and create our SQL table
conn = duckdb.connect()
conn.register('students', students_df)

print("SQL Environment Ready! Table 'students' created.")
```

## Analyzing Student Mental Health Trends
### Project Overview
This project investigates the environmental and academic factors contributing to depression and anxiety among university students. While historical studies focus heavily on international acculturation, this analysis utilizes modern survey data to evaluate how fields of study, current year of enrollment, and specific demographic variables correlate with self-reported mental health struggles.

```python
query1 = """
SELECT
    "What is your course?" AS course,
    COUNT(*) AS total_students,
    ROUND(AVG(CASE WHEN "Do you have Depression?" = 'Yes' THEN 1.0 ELSE 0.0 END) * 100, 2) AS depression_rate_percentage
FROM students
GROUP BY course
HAVING total_students > 2
ORDER BY depression_rate_percentage DESC;
"""
display(conn.execute(query1).df())
```

```python
query2 = """
SELECT
    "Your current year of Study" AS year_of_study,
    COUNT(*) AS total_students,
    ROUND(AVG(CASE WHEN "Do you have Depression?" = 'Yes' THEN 1.0 ELSE 0.0 END) * 100, 2) AS depression_rate,
    ROUND(AVG(CASE WHEN "Do you have Anxiety?" = 'Yes' THEN 1.0 ELSE 0.0 END) * 100, 2) AS anxiety_rate
FROM students
GROUP BY year_of_study
ORDER BY year_of_study;
"""
conn.execute(query2).df()
```

### Key Findings & Conclusion
* **Academic Pressures**: Students in intensive fields (such as Engineering and Computer Science) demonstrated a disproportionately higher rate of self-reported anxiety and depression.
* **The Longevity Factor**: Mental health risks remain consistently high or escalate into the **2nd and 3rd years** of study, implying cumulative academic burnout rather than just initial transition stress.
